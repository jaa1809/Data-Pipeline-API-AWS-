```python
import requests
import datetime as dt
#import pandas as pd
import sqlite3
import os
import psycopg 
import json
from dotenv import load_dotenv
load_dotenv()


dbconn = os.getenv("DBCONN")

conn = psycopg.connect (dbconn)
cur = conn.cursor()

api_key = os.getenv("API_KEY")


```

```python
# Create Customer table
cur.execute('''
DROP TABLE IF EXISTS bitcoin;
''')
```

```python
# Create Bitcoin table
cur.execute('''
CREATE TABLE IF NOT EXISTS bitcoin (
    "Date" TIMESTAMP PRIMARY KEY,
    open FLOAT,
    high FLOAT,
    low FLOAT,  
    close FLOAT,
    volume FLOAT
);
''')
```

```python
crypto = 'BTC'
btc_endpoint = "https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=EUR&apikey=" + api_key
response = requests.get(btc_endpoint)
data = response.json()
print(json.dumps(data, indent=4))
```

```python
if response.status_code == 200:
    data = response.json()  # Parse JSON
    
    # Extract the dictionary containing daily prices
    btc_time_series = data["Time Series (Digital Currency Daily)"]
    
    # Get the latest available date (sorted)
    btc_dates = sorted(btc_time_series.keys())  # Extract and sort the dates
    
    # Convert the latest date string to a datetime object
    latest_date = dt.datetime.strptime(btc_dates[0], "%Y-%m-%d")
    
    print(f"Latest available BTC data date: {latest_date}")
else:
    print(f"Error: {response.status_code}")
```
